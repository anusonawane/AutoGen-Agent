# Tool Use

````python
import re
from typing import Annotated, Literal

PasswordStrength = Literal["Weak", "Moderate", "Strong"]

def check_password_strength(password: str) -> PasswordStrength:
    if len(password) < 8:
        return "Weak"
    if not re.search(r"[A-Za-z]", password):
        return "Weak"
    if not re.search(r"[0-9]", password):
        return "Moderate"
    if not re.search(r"[!@#$%^&*(),.?\":{}|<>]", password):
        return "Moderate"
    return "Strong"
````

````python
import os
from autogen import ConversableAgent

# Create an assistant agent to suggest tool calls for checking password strength.
assistant = ConversableAgent(
    name="Assistant",# Name of the agent
    system_message="You are a helpful AI assistant. "
    "You can help with checking password strength. "
    "Return 'TERMINATE' when the task is done.", # Instructions for the assistant
    llm_config={"config_list": [{"model": "gpt-3.5-turbo", "api_key": "OPENAI_API_KEY"}]},
)

# Create a user proxy agent to execute tool calls suggested by the assistant.
user_proxy = ConversableAgent(
    name="User",
    llm_config=False,
    is_termination_msg=lambda msg: msg.get("content") is not None and "TERMINATE" in msg["content"],
    human_input_mode="NEVER",
)

# Register the check_password_strength function with the assistant agent.
# This allows the assistant to suggest calls to this function
assistant.register_for_llm(name="check_password_strength", description="Checks the strength of a password")(check_password_strength)

# Register the check_password_strength function with the user proxy agent.
# This allows the user proxy to execute calls to this function.
user_proxy.register_for_execution(name="check_password_strength")(check_password_strength)
````

````python
from autogen import register_function

# Register the check_password_strength function with both agents.
# This sets up the assistant to suggest the function and the user proxy to execute it.
register_function(
    check_password_strength,
    caller=assistant,  # The assistant agent can suggest calls to check_password_strength
    executor=user_proxy,  # The user proxy agent can execute check_password_strength calls
    name="check_password_strength",
    description="Checks the strength of a password",
)
````
